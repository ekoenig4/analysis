```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mx = np.linspace(500, 900, 100)
my = np.linspace(250, 450, 100)
partition = np.random.randint(0, 3, 100*100).reshape(100, 100)*1.0

MX, MY = np.meshgrid(mx, my)
mask = MX < 2 * MY

partition[mask] = np.nan
```

```python
plt.pcolormesh(mx, my, partition, vmin=0)
```
