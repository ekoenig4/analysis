```python
import jax
import pyhf
from pyhf.exceptions import FailedMinimization
pyhf.set_backend('jax')
import numpy as np
```

```python
norm = 2*np.sqrt(np.sum(sumw2**2))/np.sum(n_sixb)
w = pyhf.simplemodels.uncorrelated_background(
    signal=(norm*n_sixb).tolist(), bkg=n_data.tolist(), bkg_uncertainty=data.sumw2.tolist()
)

w_data = n_data.tolist() + w.config.auxdata
poi=np.linspace(0,2,21)
level=0.05

obs_limit, exp_limit = pyhf.infer.intervals.upperlimit(
          w_data, w, poi, level=level,
      )
obs_limit, exp_limit = norm*obs_limit, [ norm*lim for lim in exp_limit ]
```
